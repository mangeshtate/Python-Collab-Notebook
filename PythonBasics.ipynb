```python
def calculate_emissions(deliveries):
    EMISSION_FACTOR = 0.2  # kg CO2 per km per kg
    total_emissions = 0.0

    for distance, weight in deliveries:
        if weight == 0:
            continue  # Skip empty deliveries
        if distance > 100:
            print("Distance too far – reroute needed")
            break
        total_emissions += distance * weight * EMISSION_FACTOR

    return total_emissions

deliveries = []
n = int(input("Enter number of deliveries: "))

for i in range(n):
    distance = float(input(f"Enter distance (km) for delivery {i+1}: "))
    weight = float(input(f"Enter weight (kg) for delivery {i+1}: "))
    deliveries.append((distance, weight))

# Calculate emissions
result = calculate_emissions(deliveries)
print(f"\nTotal emissions: {result:.2f} kg CO2")
```

```python
class EnergySystem:
    def __init__(self, building_name, energy_consumption, emission_factor):
        self.building_name = building_name
        self._energy_consumption = energy_consumption
        self._emission_factor = emission_factor

    def show_details(self):
        print(f"Building: {self.building_name}")
        print(f"Energy Consumption: {self._energy_consumption} kWh")
        print(f"Emission Factor: {self._emission_factor} kg CO₂/kWh")

    def get_energy_consumption(self):
        return self._energy_consumption

    def get_emission_factor(self):
        return self._emission_factor

    def calculate_carbon_footprint(self):
        return self._energy_consumption * self._emission_factor

    def calculate_energy_saving(self):
        return self._energy_consumption * 0.10

system1 = EnergySystem("Building A", 5000, 0.45)
system1.show_details()
print(f"Carbon Footprint: {system1.calculate_carbon_footprint()} kg CO₂")
print(f"Energy Saving: {system1.calculate_energy_saving()} kWh")

class SolarEnergySystem(EnergySystem):
    def __init__(self, building_name, energy_consumption, emission_factor, solar_production):
        super().__init__(building_name, energy_consumption, emission_factor)
        self.solar_production = solar_production

    def show_details(self):
        super().show_details()
        print(f"Solar Production: {self.solar_production} kWh")

    def calculate_carbon_footprint(self):
        net_consumption = self.get_energy_consumption() - self.solar_production
        return net_consumption * self.get_emission_factor()

solar_system = SolarEnergySystem(" Solar Building", 5000, 0.45, 1500)
solar_system.show_details()
print(f"Net Carbon Footprint: {solar_system.calculate_carbon_footprint()} kg CO₂")
```

```python
# Sustainable Resource Tracker
class Building:
    def __init__(self, name, btype, electricity, water, waste, occupants=0):
        self.name = name
        self.btype = btype  # Residential, Commercial, Industrial
        self.electricity = electricity  # kWh
        self.water = water  # liters
        self.waste = waste  # kg
        self.occupants = occupants  # Only for residential

    def carbon_footprint(self):
        # Assume 0.85 kg CO2 per kWh
        return self.electricity * 0.85

    def water_usage_per_person(self):
        if self.btype == "Residential" and self.occupants > 0:
            return self.water / self.occupants
        return None

    def waste_efficiency_rating(self):
        if self.btype == "Residential":
            return "Good" if self.waste < 200 else "Poor"
        elif self.btype == "Commercial":
            return "Efficient" if self.waste < 500 else "Inefficient"
        elif self.btype == "Industrial":
            return "Sustainable" if self.waste < 1000 else "Unsustainable"
        return "Unknown"

    def meets_eco_targets(self):
        # Custom eco target logic
        if self.carbon_footprint() < 1000 and self.waste_efficiency_rating() in ["Good", "Efficient", "Sustainable"]:
            return True
        return False


class ResourceTracker:
    def __init__(self):
        self.buildings = []

    def add_building(self, building):
        self.buildings.append(building)

    def average_usage(self):
        avg_electricity = sum(b.electricity for b in self.buildings) / len(self.buildings)
        avg_water = sum(b.water for b in self.buildings) / len(self.buildings)
        avg_waste = sum(b.waste for b in self.buildings) / len(self.buildings)
        return avg_electricity, avg_water, avg_waste

    def identify_above_average(self):
        avg_elec, avg_water, avg_waste = self.average_usage()
        above_avg = []
        for b in self.buildings:
            if b.electricity > avg_elec or b.water > avg_water or b.waste > avg_waste:
                above_avg.append(b.name)
        return above_avg

    def display_report(self):
        print("\n--- Sustainable Resource Tracker Report ---\n")
        for b in self.buildings:
            print(f"Building: {b.name} ({b.btype})")
            print(f"  Electricity: {b.electricity} kWh")
            print(f"  Water: {b.water} L")
            print(f"  Waste: {b.waste} kg")
            print(f"  Carbon Footprint: {b.carbon_footprint():.2f} kg CO2")
            if b.btype == "Residential":
                print(f"  Water per person: {b.water_usage_per_person():.2f} L/person")
            print(f"  Waste Efficiency: {b.waste_efficiency_rating()}")
            print(f"  Meets Eco Targets: {'Yes' if b.meets_eco_targets() else 'No'}\n")

        print("Above Average Resource Users:", self.identify_above_average())



if __name__ == "__main__":
    tracker = ResourceTracker()
    # Add some buildings
    tracker.add_building(Building("Green Apartments", "Residential", electricity=800, water=3000, waste=150, occupants=15))
    tracker.add_building(Building("Tech Mall", "Commercial", electricity=5000, water=7000, waste=600))
    tracker.add_building(Building("Steel Factory", "Industrial", electricity=12000, water=15000, waste=1200))
    # Generate sustainability report
    tracker.display_report()
```

```python
# Sustainable Resource Tracker (Updated with given formulas)

class Building:
    def __init__(self, name, btype, electricity, water, waste, occupants=0, employees=0, units=0):
        self.name = name
        self.btype = btype  # Residential, Commercial, Industrial
        self.electricity = electricity  # kWh
        self.water = water  # liters
        self.waste = waste  # kg
        self.occupants = occupants  # Only for residential
        self.employees = employees  # Only for commercial
        self.units = units  # Only for industrial

    def carbon_footprint(self):
        # Emission Factor: 0.92 kg CO2/kWh
        return self.electricity * 0.92

    def water_usage_per_person(self):
        if self.btype == "Residential" and self.occupants > 0:
            return self.water / self.occupants
        return None

    def waste_efficiency_rating(self):
        if self.btype == "Residential" and self.occupants > 0:
            return "Efficient" if (self.waste / self.occupants) <= 20 else "Inefficient"
        elif self.btype == "Commercial" and self.employees > 0:
            return "Efficient" if (self.waste / self.employees) <= 10 else "Inefficient"
        elif self.btype == "Industrial" and self.units > 0:
            return "Efficient" if (self.waste / self.units) <= 0.5 else "Inefficient"
        return "Unknown"

    def meets_eco_targets(self):
        # Custom target: Low carbon + Efficient waste usage + Recommended water usage
        carbon_ok = self.carbon_footprint() < 1000
        waste_ok = self.waste_efficiency_rating() == "Efficient"
        water_ok = True

        if self.btype == "Residential" and self.occupants > 0:
            water_ok = self.water_usage_per_person() <= 4050  # Recommended limit

        return carbon_ok and waste_ok and water_ok


class ResourceTracker:
    def __init__(self):
        self.buildings = []

    def add_building(self, building):
        self.buildings.append(building)

    def average_usage(self):
        avg_electricity = sum(b.electricity for b in self.buildings) / len(self.buildings)
        avg_water = sum(b.water for b in self.buildings) / len(self.buildings)
        avg_waste = sum(b.waste for b in self.buildings) / len(self.buildings)
        return avg_electricity, avg_water, avg_waste

    def identify_above_average(self):
        avg_elec, avg_water, avg_waste = self.average_usage()
        above_avg = []
        for b in self.buildings:
            if b.electricity > avg_elec or b.water > avg_water or b.waste > avg_waste:
                above_avg.append(b.name)
        return above_avg

    def display_report(self):
        print("\n--- Sustainable Resource Tracker Report ---\n")
        for b in self.buildings:
            print(f"Building: {b.name} ({b.btype})")
            print(f"  Electricity: {b.electricity} kWh")
            print(f"  Water: {b.water} L")
            print(f"  Waste: {b.waste} kg")
            print(f"  Carbon Footprint: {b.carbon_footprint():.2f} kg CO2")

            if b.btype == "Residential":
                print(f"  Water per person: {b.water_usage_per_person():.2f} L/person (Recommended ≤ 4050)")

            print(f"  Waste Efficiency: {b.waste_efficiency_rating()}")
            print(f"  Meets Eco Targets: {'Yes' if b.meets_eco_targets() else 'No'}\n")

        print("Above Average Resource Users:", self.identify_above_average())
        print("\n------------------------------------------\n")


# ---------------- Example Usage ---------------- #
if __name__ == "__main__":
    tracker = ResourceTracker()

    # Residential: 15 people
    tracker.add_building(Building("Green Apartments", "Residential", electricity=800, water=50000, waste=250, occupants=15))

    # Commercial: 50 employees
    tracker.add_building(Building("Tech Mall", "Commercial", electricity=5000, water=70000, waste=400, employees=50))

    # Industrial: 3000 production units
    tracker.add_building(Building("Steel Factory", "Industrial", electricity=12000, water=150000, waste=1200, units=3000))

    # Generate sustainability report
    tracker.display_report()
```

```python
class Building:
    def __init__(self, name, electricity, water, waste):
        self.name = name
        self.electricity = electricity  # kWh
        self.water = water              # liters
        self.waste = waste              # kg

    def carbon_footprint(self):
        return self.electricity * 0.5

    def water_usage(self):
        # Only Residential overrides this
        return None

    def waste_efficiency(self):
        return "Not defined"

    def display_info(self):
        print(f"\nBuilding: {self.name}")
        print(f"  Electricity: {self.electricity} kWh")
        print(f"  Water: {self.water} L")
        print(f"  Waste: {self.waste} kg")
        print(f"  Carbon Footprint: {self.carbon_footprint()} kg CO2")
        if self.water_usage() is not None:
            print(f"  Water usage per person: {self.water_usage():.2f} L/person")
        print(f"  Waste Efficiency: {self.waste_efficiency()}")


class Residential(Building):
    def __init__(self, name, electricity, water, waste, num_people):
        super().__init__(name, electricity, water, waste)
        self.num_people = num_people

    def water_usage(self):
        return self.water / self.num_people if self.num_people > 0 else 0

    def waste_efficiency(self):
        # Example: less than 30 kg per person = efficient
        return "Efficient" if (self.waste / self.num_people) < 30 else "Needs Improvement"


class Commercial(Building):
    def waste_efficiency(self):
        # Example: < 500 kg/month is efficient
        return "Efficient" if self.waste < 500 else "High Waste"


class Industrial(Building):
    def waste_efficiency(self):
        # Example: < 2000 kg/month is efficient
        return "Efficient" if self.waste < 2000 else "Excessive Waste"


# ----------------Main Tracker----------------
buildings = [
    Residential("Green Apartments", 1200, 30000, 800, 50),
    Commercial("City Mall", 5000, 20000, 1200),
    Industrial("Steel Factory", 20000, 50000, 3000),
    Residential("Sunny Homes", 800, 15000, 1000, 40)
]

for b in buildings:
    b.display_info()

# --Analysis--
# 1. Buildings with above-average electricity usage
avg_electricity = sum(b.electricity for b in buildings) / len(buildings)
print("\n Buildings with above-average electricity usage:")
for b in buildings:
    if b.electricity > avg_electricity:
        print(f" - {b.name}")

# 2. Buildings that meet eco-efficiency targets
print("\n Eco-efficient buildings:")
for b in buildings:
    if "Efficient" in b.waste_efficiency():
        print(f" - {b.name}")
```
